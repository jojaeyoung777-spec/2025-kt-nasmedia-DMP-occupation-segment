```python
import pandas as pd

df = match
# 1. 알파벳 코드 -> 산업분류명 매핑 딕셔너리 정의
code_to_name = {
    "A": "농업, 임업 및 어업(01~03)",
    "B": "광업(05~08)",
    "C": "제조업(10~34)",
    "D": "전기, 가스, 증기 및 공기조절 공급업(35)",
    "E": "수도, 하수 및 폐기물 처리, 원료 재생업(36~39)",
    "F": "건설업(41~42)",
    "G": "도매 및 소매업(45~47)",
    "H": "운수 및 창고업(49~52)",
    "I": "숙박 및 음식점업(55~56)",
    "J": "정보통신업(58~63)",
    "K": "금융 및 보험업(64~66)",
    "L": "부동산업(68)",
    "M": "전문, 과학 및 기술 서비스업(70~73)",
    "N": "사업시설 관리, 사업 지원 및 임대 서비스업(74~76)",
    "O": "공공행정, 국방 및 사회보장 행정(84)",
    "P": "교육 서비스업(85)",
    "Q": "보건업 및 사회복지 서비스업(86~87)",
    "R": "예술, 스포츠 및 여가관련 서비스업(90~91)",
    "S": "협회 및 단체, 수리 및 기타 개인 서비스업(94~96)",
    "T": "가구 내 고용활동 및 기타 자가소비 생산활동(97~98)",
    "U": "국제 및 외국기관(99)",
    "Z": "기타/미분류"
}

# 2. 매핑 적용하여 새로운 컬럼(depth1_nm) 생성
# 만약 딕셔너리에 없는 코드가 있다면 NaN이 됩니다. 이를 막기 위해 .get()이나 fillna를 쓸 수도 있습니다.
df['depth1_nm'] = df['corp_depth1_cd'].map(code_to_name).fillna("기타/미분류")

# 3. 산업분류명(depth1_nm)별 adid 수 집계
# adid가 중복될 수 있으므로 nunique()를 사용하여 고유 유저 수를 셉니다.
# (단순 행 개수가 필요하면 count()를 사용하세요)
result = df.groupby('depth1_nm')['adid'].nunique().reset_index()

# 4. 결과 컬럼명 변경 (보기 좋게)
result.columns = ['산업분류명', '유저수(ADID)']

# 5. 유저 수 많은 순서대로 정렬
result = result.sort_values(by='유저수(ADID)', ascending=False)

# 결과 확인
result
```
